Now we have the shortened list of columns after feature selection stp is over. Now we shall do the GridSearchCV/RandomizedSearchCV  for the XGBoost algorithm

```python
#Apply a GridSearchCV for XGBoost algorithm
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

# param_grid = {
#     'n_estimators': [100, 200, 300],
#     'learning_rate': [0.01, 0.1, 0.2],
#     'max_depth': [3, 4, 5],
#     'subsample': [0.8, 0.9, 1.0],
#     'colsample_bytree': [0.8, 0.9, 1.0],
#     'gamma': [0, 0.1, 0.2],
# }
param_dist = {
    'n_estimators': randint(100, 500),  # Uniform distribution between 100 and 500
    'learning_rate': uniform(0.01, 0.2), # Uniform distribution between 0.01 and 0.2
    'max_depth': randint(3, 7),        # Uniform distribution between 3 and 7
    'subsample': uniform(0.7, 0.3),      # Uniform distribution between 0.7 and 1.0 (0.7 + 0.3)
    'colsample_bytree': uniform(0.7, 0.3), # Uniform distribution between 0.7 and 1.0 (0.7 + 0.3)
    'gamma': uniform(0, 0.5)              # Uniform distribution between 0 and 0.5
}

# Create RandomizedSearchCV object
randomized_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_dist,
    n_iter=5,          # Number of random combinations to try
    scoring='neg_mean_squared_error',
    cv=5,
    verbose=2,
    random_state=42,    # Set random_state for reproducibility
    n_jobs=-1           # Use all available cores
)
```

```python
# Fit the randomized search to your training data
%%timeit
randomized_search.fit(X_train_selected, y_train)
```

```python
# Get the best parameters and best score
print("Best parameters:", randomized_search.best_params_)
print("Best score:", randomized_search.best_score_)
```

Now we have the best fit XGBoost model. and Now we can use it to fit the real train data.

```python
# Get the best model from RandomizedSearchCV
best_xgb_model = randomized_search.best_estimator_

# Fit the best model to the entire training data
%timeit
best_xgb_model.fit(X_train_selected, y_train)
```

Now evaluate against validation set

```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the validation set
predictions = best_xgb_model.predict(X_valid_selected)

# Calculate evaluation metrics
mse = mean_squared_error(y_valid, predictions)
rmse = np.sqrt(mse)
r2 = r2_score(y_valid, predictions)

# Print the results
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R2): {r2}")
```

```python
from sklearn.pipeline import Pipeline
from feature_engine.imputation import CategoricalImputer
from feature_engine.imputation import MeanMedianImputer
from feature_engine.encoding import RareLabelEncoder
import category_encoders as ce
from sklearn.preprocessing import MinMaxScaler
from feature_engine.wrappers import SklearnTransformerWrapper
from sklearn.feature_selection import SelectFromModel
from feature_engine.outliers import OutlierTrimmer
import joblib

# shrtnd_list=pd.DataFrame(X_train_selected, columns=selected_features)
# #shrtnd_list.dtypes[X_train_selected.dtypes == 'object'].index.tolist()
# shrtnd_list_of_cat_columns=shrtnd_list.dtypes[shrtnd_list.dtypes == 'object'].index.tolist()
# shrtnd_list_of_num_columns=pd.DataFrame(X_train_selected, columns=selected_features).select_dtypes(include=['int64', 'float64']).columns.tolist()

categorical_columns = ['Alley',
 'LandSlope',
 'Neighborhood',
 'Condition1',
 'RoofMatl',
 'MasVnrType',
 'ExterQual',
 'BsmtQual',
 'BsmtExposure',
 'CentralAir',
 'KitchenQual',
 'GarageType',
 'GarageFinish',
 'GarageCond',
 'PavedDrive',
 'SaleType',
 'SaleCondition']
num_columns=['LotFrontage',
 'LotArea',
 'OverallQual',
 'OverallCond',
 'YearBuilt',
 'YearRemodAdd',
 'MasVnrArea',
 'BsmtFinSF1',
 'TotalBsmtSF',
 '1stFlrSF',
 '2ndFlrSF',
 'GrLivArea',
 'BsmtFullBath',
 'BsmtHalfBath',
 'FullBath',
 'KitchenAbvGr',
 'TotRmsAbvGrd',
 'Fireplaces',
 'GarageCars',
 'GarageArea',
 'ScreenPorch',
 'PoolArea',
 'MoSold']

# Load the joblib file
best_xgb_model_loaded = joblib.load('/content/drive/MyDrive/Colab Notebooks/best_xgb_model.joblib')

```

```python
# Assuming we defined our preprocessing steps (rare_encoder, cat_imputer_frequent, num_imputer_mean, etc.) and your XGBoost model (xgb)

# Create a pipeline
pipeline = Pipeline([

    # Imputation for categorical variables
    # ('cat_imputer', CategoricalImputer(imputation_method='frequent', variables=list_of_categorical_columns)),
    ('cat_imputer', CategoricalImputer(imputation_method='frequent', variables=categorical_columns)),

    # Imputation for numerical variables
    # ('num_imputer', MeanMedianImputer(imputation_method='mean', variables=X.select_dtypes(include=['int64','float64']).columns.tolist())),
    ('num_imputer', MeanMedianImputer(imputation_method='mean', variables=num_columns)),

    # Imputation for categorical variables
    # ('outlier_trimmer', OutlierTrimmer(capping_method='gaussian', tail='both', fold=2.5, variables=X.select_dtypes(include=['int64','float64']).columns.tolist())),
    ('outlier_trimmer', OutlierTrimmer(capping_method='gaussian', tail='both', fold=2.5, variables=num_columns)),

    # Rare label encoding
    # ('rare_encoder', RareLabelEncoder(tol=0.02, n_categories=20, variables=list_of_categorical_columns)),
    ('rare_encoder', RareLabelEncoder(tol=0.02, n_categories=4, variables=categorical_columns)),

    # Target encoding
    ('target_encoder', ce.TargetEncoder(cols=categorical_columns)),

    # Scaling
    ('scaler', SklearnTransformerWrapper(transformer=MinMaxScaler())),

    # XGBoost model
    ('xgb_model', best_xgb_model_loaded)
])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df_pl=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/housing_train.csv')

df_pl.drop_duplicates(inplace=True)

selected_features=categorical_columns+num_columns+['SalePrice']
df_pl_sel=df_pl[selected_features]

X_pl_sel,y_pl_sel=df_pl_sel.drop(columns=['SalePrice']),df_pl_sel['SalePrice']

# print(X_pl_sel)
# print(y_pl_sel)

#X_pl_train,y_pl_train=train_test_split(X_pl,y_pl,test_size=0.2,random_state=42)
X_pl_train, X_pl_valid, y_pl_train, y_pl_valid = train_test_split(X_pl_sel, y_pl_sel, test_size=0.2, random_state=42)

print(X_pl_train.shape)
print(y_pl_train.shape)
print(X_pl_valid.shape)
print(y_pl_valid.shape)
```
