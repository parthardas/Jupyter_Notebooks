### Deciding how to fill null values. we could have taken into consideration several ways but will majorly 2 of them, which makes sense over here.

### Points to be Noted: 
1.) Dropping missing values: If the number of missing values is relatively small compared to the total dataset and they do not significantly impact the analysis or modeling, you can simply drop the rows or columns containing missing values. However, this approach should be used with caution, as it can result in a loss of valuable information.

2.) Mean/Median/Mode imputation: For numerical features, you can fill missing values with the mean, median, or mode of the available data. This approach assumes that the missing values are missing at random and that the non-missing values provide a representative estimate for the missing values. Use mean imputation for normally distributed data and median imputation for skewed data. Mode imputation is suitable for categorical features.

### Here since the data in both the columns, age & bmi have skewness between the range -0.5 to +0.5, so we can say they are fairly symmetrical and hence we proceed with filling null values with mean.

#### Based on the three box plots of the numeric columns, it appears that there are outliers only in the "claims" column. Since the "claims" column has a skewed distribution, outlier checking will be performed using the IQR method.

```python
# Outlier Detection For Skewed Distribution
def skewed_dis(df, col):
    IQR = df[col].quantile(0.75) - df[col].quantile(0.25)
    lower_bound = df[col].quantile(0.25) - (IQR * 1.5)
    upper_bound = df[col].quantile(0.75) + (IQR * 1.5)
    return df[(df[col] < lower_bound) | (df[col] > upper_bound)] 
```

```python
outliers = skewed_dis(df_train, 'claim')
print(f"Outlier percentage: {(len(outliers) / len(df_train) * 100):.1f}%")
outliers
```

## Pipeline

### In this section, a pipeline will be created to perform scaling using StandardScaler(), encoding using OneHotEncoder(), and then dimensionality reduction using PCA() with n_component = 0.95. This means that the dimensions will be reduced while retaining 95% of the information from the data.

```python
# Numeric Features Scaling
num_pipe = Pipeline([('scaling', StandardScaler())])

# Categorical Features Encoding
cat_pipe = Pipeline([('encode', OneHotEncoder(handle_unknown='ignore'))])

# Preprocessing Pipeline
preprocess = ColumnTransformer([
    ('num',num_pipe, X_train.select_dtypes(include=['float64', 'int64']).columns),
    ('cat',cat_pipe, X_train.select_dtypes(include=['object']).columns)
])
```

#### Creating a pipeline for Polynomial Regression and Extreme Gradient Boosting Regressor models.

```python
poly_lr_pipe = Pipeline([('preprocessing', preprocess), ('pca', PCA(n_components=0.95)), ('poly', PolynomialFeatures()), ('model', LinearRegression())])
```

#### Creating a dictionary that contains hyperparameters for the Polynomial Regression and Extreme Gradient Boosting Regressor models.

```python
poly_params = {'poly__degree': [3, 4, 5, 6]}
```

```python
poly = GridSearchCV(poly_lr_pipe, param_grid=poly_params, cv=2, n_jobs=-1)
```
